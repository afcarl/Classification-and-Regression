Demonstrating the impact of using higher order basis functions. In this notebook we are using the polynomial fitting function available in the numpy library.

```python
import numpy as np
import numpy.polynomial.polynomial as poly
```

```python
#generate some data
x = np.arange(20)
w = np.array([-3.8,0.11])
y = w[0] + w[1]*x
sigma2 = 0.4
y = y + np.random.normal(0,sqrt(sigma2),x.shape[0])

plt.scatter(x,y)
```

```python
# p controls the degree of the polynomial
plt.scatter(x,y)
for p in range(15):
    x_new = np.linspace(x[0], x[-1], num=len(x)*10)
    w = poly.polyfit(x, y, p)
    print w
    ffit = poly.polyval(x_new, w)
    plt.plot(x_new, ffit)
```
